```python
def readcsvfile(filename):
    file_open=open(filename,'r')
    lines_data=[]
    for line in iter(file_open):
        if 'subject' in line:
            continue
        line=line.rstrip()
        lines_data.append(line.split(','))
    return lines_data
```
