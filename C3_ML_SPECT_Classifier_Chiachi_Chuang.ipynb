SPECT Heart Data Set

https://archive.ics.uci.edu/ml/datasets/SPECT+Heart

Data on cardiac Single Proton Emission Computed Tomography (SPECT) images. 
Each patient classified into two categories: normal and abnormal.

23 Attributes:

Attribute Information:
1. OVERALL_DIAGNSIS: 0,1 (class attribute, binary)
2.- 23. F1-F22: 0,1 (the partial diagnosis 1-22, binary)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readFile(filename):
    contents=[]
    file1=open(filename,'r')
    for line in iter(file1):
        line=line.strip()
        contents.append(line.split(','))
    return contents    
```

```python
trainDS=readFile('SPECT.train')
testDS=readFile('SPECT.test')
```

```python
print(trainDS[:3])
```

```python
header1=["class","F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","F12","F13","F14","F15","F16","F17","F18","F19","F20","F21","F22"]
traindf = pd.DataFrame(trainDS,columns=header1)
```

```python
print(traindf.head())
```

```python
traindf.describe()
```

```python
dataset = np.array(trainDS)
X1 = dataset[:,1:]
Y1 = dataset[:,0]
```

```python
# Change target values to integers
X2 = X1.astype(np.int)
Y2 = Y1.astype(np.int)
```

```python
ds_test = np.array(testDS)
X5 = ds_test[:,1:]
Y5 = ds_test[:,0]
Xt = X5.astype(np.int)
Yt = Y5.astype(np.int)
```

```python
X_header=["F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","F12","F13","F14","F15","F16","F17","F18","F19","F20","F21","F22"]
df_Xy = pd.DataFrame(X2,columns=X_header)
df_Xy["class"]=Y2
```

```python
df_Xy.describe()
```

```python
fig = plt.figure(figsize = (16,4))
df_Xy.boxplot()
```

```python
df_Xy.corr()
```

```python
# class: normal -- 0
# class: abnormal -- 1
# F8, F13, F16 seems has higher correclation with class
# F5 has lowest correlation with class
features = X2
target = Y2

f, ax = plt.subplots(1)
for i in np.unique(target):
    mask = (target == i)
    #print(features[mask,0])
    plt.scatter(features[mask, 12], features[mask, 4], label=i)
ax.legend()
plt.xlabel('F13')
plt.ylabel('F5')
plt.show()
```

```python
# Logistic Regression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X2,Y2)
print(model)
# make predictions
expected = Y2
predicted = model.predict(X2)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
def plot_cm(cm):   
    classes = ["normal","abnormal"]
    df_cm = pd.DataFrame(cm, index=classes, columns=classes)
    fig = plt.figure(figsize=(5,5))
    heatmap = sns.heatmap(df_cm, annot=True, fmt="d",cmap="YlGnBu",linecolor="blue",linewidths=1.0)
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=10)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=10)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
plot_cm(cm)
```

```python
# Logistic Regression
# make predictions on test data
expected = Yt
predicted = model.predict(Xt)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
plot_cm(cm)
```
